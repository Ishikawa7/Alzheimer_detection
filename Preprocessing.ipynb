```python
import os
import numpy as np
import pickle as pkl
from PIL import Image
import random
```

```python
dir_train = "Train/"
dir_test = "Test/"

dict_train = {}
dict_test = {}

dict_train_count = {}
dict_test_count = {}
```

```python
sum_train = 0
sum_test = 0
for folder in os.listdir(dir_train):
    count_train = len(os.listdir(dir_train+folder))
    count_test = len(os.listdir(dir_test+folder))
    sum_train += count_train
    sum_test += count_test
    dict_train_count[folder] = count_train
    dict_test_count[folder] = count_test
dict_train_count["Total"] = sum_train
dict_test_count["Total"] = sum_test
print("Train Count")
print(dict_train_count)
print("Test Count")
print(dict_test_count)
```

```python
for key in dict_train_count.keys():
    dict_train_count[key] = sum_train/dict_train_count[key]
dict_train_count
# normalize with min-max
max_train = max(dict_train_count.values())
min_train = min(dict_train_count.values())
for key in dict_train_count.keys():
    dict_train_count[key] = (dict_train_count[key]-(min_train))/((max_train)-(min_train))
dict_train_count
```

```python
for folder in os.listdir(dir_train):
    for image in os.listdir(dir_train+folder):
        image_file = Image.open(dir_train+folder+"/"+image)
        np_image = np.array(image_file)
        np_image = np_image/np_image.max()
        image_file.close()
        dict_image = {}
        dict_image["label"] = folder
        dict_image["class_weight"] = dict_train_count[folder]
        dict_image["image"] = np_image
        dict_train[image] = dict_image
        # shuffle dict
dict_train_list = list(dict_train.items())
random.shuffle(dict_train_list)
dict_train_shuffled = dict(dict_train_list)
# save dict
with open('train_dict.pkl', 'wb') as f:
    pkl.dump(dict_train_shuffled, f)
dict_train_shuffled
```

```python
for folder in os.listdir(dir_test):
    for image in os.listdir(dir_test+folder):
        image_file = Image.open(dir_test+folder+"/"+image)
        np_image = np.array(image_file)
        np_image = np_image/np_image.max()
        image_file.close()
        dict_image = {}
        dict_image["label"] = folder
        dict_image["image"] = np_image
        dict_image["class_weight_train"] = dict_train_count[folder]
        dict_test[image] = dict_image
# shuffle dict
dict_test_list = list(dict_test.items())
random.shuffle(dict_test_list)
dict_test_shuffled = dict(list(dict_test.items()))
# save dict
with open('test_dict.pkl', 'wb') as f:
    pkl.dump(dict_test_shuffled, f)
dict_test_shuffled
```
